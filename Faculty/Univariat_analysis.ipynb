Frecuencia Absoluta

```r
x <- c(1, 1, 2, 3.5, 4, 4, 4, 5, 5, 6.5, 7, 7.4, 8, 13, 14.2)
```

```r
mean(x)
```

```r
median(x)
```

```r
max(x) - min(x)
```

```r
IQR(x) #intercuantil
```

```r
sd(x) #standar desviation
```

```r
table(x) #frecuency table, they 1 ocurrs two times, they 5 two times, they 2 one time
```

```r
hist(x)
```

```r
plot(density(x))
```
